```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
```

```python
for i in train.columns:
    if train[i].dtype == object:
        train[i] = label.fit_transform(train[i])


for i in test.columns:
    if test[i].dtype == object:
        test[i] = label.fit_transform(test[i])
```

```python
X = train.drop('Exited', axis=1)
y = train['Exited']
X_test = test
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
import xgboost as xgb
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
params = {"booster": "gbtree",
          "max_depth": 3,
          "gamma": 0,
          "learning_rate": 0.04538451353216046,
          "n_estimators": 1171,
          "min_child_weight": 13,
          "subsample": 0.6578720167306905,
          "reg_alpha": 0.4622943878867952,
          "reg_lambda": 0.6211309481623338,
          "colsample_bytree": 0.9634723040072966,
          "colsample_bylevel": 0.7985625445322195,
          "colsample_bynode": 0.49814271378837327,
          "random_state": 42,
          "objective": "binary:logistic",
          "n_jobs": -1}

model_xgb = xgb.XGBClassifier(**params)
model_xgb.fit(X_train, y_train)
```

```python
# print(classification_report(y_test, y_pred))
print(accuracy_score(y_test, y_pred))
```
